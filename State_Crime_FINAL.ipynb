```python
new_state_crime_df['state_lower'] = new_state_crime_df['State'].str.lower()
updated_income['state_lower'] = updated_income['State'].str.lower()
```

```python
# Create DF for Under 18 vs Household Median Income
age_by_income = pd.merge(new_state_crime_df, updated_income, on ='state_lower')
age_by_income.head()
```

```python
# TOTAL CASES OVER 18
age_by_income["Total Cases Over 18"] = (age_by_income["Total Cases_All"] - age_by_income["Total Cases_Under"])
age_by_income.columns
```

```python
# DATAFRAME WITH TOTAL CASES UNDER 18 AND OVER 18
age_under_over18 = age_by_income.loc[:, ["state_lower","Total Cases_Under","Total Cases Over 18","2016 Median Household Income"]]
age_under_over18["2016 Median Household Income"] = age_under_over18["2016 Median Household Income"].str.replace(",","").astype(float)
age_under_over18 = age_under_over18.rename(columns={"state_lower": "States", "Total Cases_Under": "Total Cases Under 18"})
age_under_over18.head()
```

```python
# DF with = UNDER 18 and INCOME - Ascending
age_under18_df = age_under_over18.loc[:, ["States", "Total Cases Under 18", "2016 Median Household Income"]]
age_under18_df.sort_values(by='2016 Median Household Income', ascending=True).head()
```

```python
# DF with = UNDER 18 and INCOME - Descending
age_under18_df = age_under_over18.loc[:, ["States", "Total Cases Under 18", "2016 Median Household Income"]]
age_under18_df.sort_values(by='2016 Median Household Income', ascending=False).head()
```

```python
# DF with = OVER 18 and INCOME - Ascending
age_over18_df = age_under_over18.loc[:, ["States", "Total Cases Over 18", "2016 Median Household Income"]]
age_over18_df.sort_values(by='2016 Median Household Income', ascending=True)
```

```python
# DF with = OVER 18 and INCOME - Descending
age_over18_df = age_under_over18.loc[:, ["States", "Total Cases Over 18", "2016 Median Household Income"]]
age_over18_df.sort_values(by='2016 Median Household Income', ascending=False).head()
```

```python
median_income = age_over18_df["2016 Median Household Income"].median()
print(f"The Median Household Income for USA is {median_income}.")
```

```python
# BAR GRAPH - UNDER 18, OVER 18, MEDIAN HOUSEHOLD
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = age_under_over18.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(age_under_over18["States"], rotation=45)

plt.show()
```

```python
new_age_income = age_by_income.loc[:, ['state_lower', 'Total Cases_Under', '2016 Median Household Income', '2016 Estimated Population_Under','Total Cases_All']]
new_age_income = new_age_income.rename(columns={
    "state_lower": "State",
    "Total Cases_Under": "Total Cases Under 18",
    "Total Cases_All": "Total Cases"
})

new_age_income["2016 Median Household Income"] = new_age_income["2016 Median Household Income"].str.replace(",","").astype(float)
new_age_income.head()
```

```python
new_age_income.sort_values(by='Total Cases Under 18', ascending=False).head()
```

```python
new_age_income.dtypes
```

```python
# PERCENT OF UNDER 18 CASES BY TOTAL POPULATION
new_age_income["Total Cases Over 18"] = age_over18_df["Total Cases Over 18"]
new_age_income["% under 18"] = (new_age_income['Total Cases Under 18'] / new_age_income['2016 Estimated Population_Under']) * 100
#new_age_income["% under 18"] = new_age_income["% under 18"].map("{:,.2%}".format)
new_age_income["% over 18"] = (new_age_income['Total Cases Over 18'] / new_age_income['2016 Estimated Population_Under']) * 100
#new_age_income["% over 18"] = new_age_income["% over 18"].map("{:,.2%}".format)
new_age_income["Total Cases"] = new_age_income["Total Cases Over 18"] + new_age_income["Total Cases Under 18"]
new_age_income["% Total Cases"] = (new_age_income['Total Cases'] / new_age_income['2016 Estimated Population_Under']) * 100

new_age_income.sort_values(by='2016 Median Household Income', ascending=False)

#Reorganize Columns
new_age_income = new_age_income[["State","Total Cases Under 18","% under 18","Total Cases Over 18","% over 18",
                                "2016 Estimated Population_Under","2016 Median Household Income","Total Cases", "% Total Cases"]]


new_age_income
```

```python
# TOTAL CASES by POPULATION
x_values = new_age_income["2016 Median Household Income"]
y_values = new_age_income["% Total Cases"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('2016 Median Household Income')
plt.ylabel('Total Cases Crime Rate (%)')
plt.show
```

```python
print(f"There is a strong negative correlation between crime rate and median household income overall.")
print("As income increases, crime rate decreases.")
```

```python
x_values = new_age_income["2016 Median Household Income"]
y_values = new_age_income["% under 18"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('2016 Median Household Income')
plt.ylabel('Under 18 Crime Rate (%)')
plt.show
```

```python
print("There is no profound correlation between household income and underaged crime.")
print("The underage population seems to show no change as household income increases.")
```

```python
x_values = new_age_income["2016 Median Household Income"]
y_values = new_age_income["% over 18"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.scatter(x_values, y_values)
plt.xlabel('2016 Median Household Income')
plt.ylabel('Over 18 Crime Rate (%)')
plt.show
```
